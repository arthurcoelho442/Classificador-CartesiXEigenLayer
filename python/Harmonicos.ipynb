# **Calculo da Transformada de Fourier**

```python
#@title Transformação dos dados para uma onda

plt.figure(figsize=(12,5))
plt.title("Classe 4", fontsize = 20)
plt.ylabel("Corrente (A)",fontsize = 15)
plt.xlabel('Tempo(s)',  fontsize = 15)
plt.grid(True)
plt.plot(vetorIntervalos, dados)
plt.show()

from scipy import signal
# Faz a transformada rápida de fourrier
fft = np.fft.fft(dados)

# Faz a transformada rápida de fourrier
fast = np.fft.fftfreq(amostras, T)

# Seleciona a primeira metade da sequência e normaliza as amplitudes pelo número de amostras (divide por N)
freqs = fast[:amostras//2]
```

```python
#@title
amplet = np.abs(fft)[:amostras//2] / amostras

#observar as diferenças de dB até 15 considerar

# Prepara o gráfico
plt.figure(figsize=(12,5)) 
plt.title("FFT do sinal", fontsize = 20)
plt.ylabel("Amplitude (A)",fontsize = 15)
plt.xlabel("Frequência (Hz)", fontsize = 15)
plt.grid(True)
plt.xlim(0, 1)

plt.plot(freqs, amplet)

# Plot peaks.
qtd_Peaks = 5
pontos = signal.argrelextrema(amplet, np.greater)
pontos = pontos[0]

peak_x = list(np.abs(freqs[pontos]))
peak_y = list(np.abs(amplet[pontos]))

plt.plot(peak_x[:qtd_Peaks], peak_y[:qtd_Peaks], marker='o', linestyle='dashed', color='green', label="Peaks")
plt.legend(fontsize = 15)

plt.show()
# Printa as Amplitudes emcontradas
a = [round(aux*10**3,3) for aux in peak_y[:qtd_Peaks]]
f = [round(aux*10**3,3) for aux in peak_x[:qtd_Peaks]]
print("Valor das Amplitudes: {} [mA]".format(a))
print("Valor das Frequencias: {} [mHz]".format(f))
```
